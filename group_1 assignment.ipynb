```python
#LOOPS
movies = [
    {"title": "Inception", "ratings": [5, 4, 5, 5]},
    {"title": "Titanic", "ratings": [4, 4, 5, 4]},
    {"title": "Avatar", "ratings": [5, 5, 5, 4]},
    {"title": "Interstellar", "ratings": [4, 5, 4, 5]}
]

# 1️⃣ Calculate average rating (rounded to 2 decimals)
for movie in movies:
    avg = sum(movie["ratings"]) / len(movie["ratings"])
    movie["average"] = round(avg, 2)

# 2️⃣ Find movies with average ≥ 4.5
top_movies = [movie["title"] for movie in movies if movie["average"] >= 4.5]

# 3️⃣ Find highest-rated movie
highest_rated = max(movies, key=lambda x: x["average"])["title"]

print("Movies with averages:", movies)
print("Top Movies:", top_movies)
print("Highest Rated:", highest_rated)
```
